```python
# 1. Imports
import pandas as pd
import numpy as np
import os

# 2. Create project folder
os.makedirs("pollution_monitoring", exist_ok=True)

# 3. Generate synthetic sensor data (5000+ rows)
np.random.seed(42)
cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bangalore']
dates = pd.date_range('2023-01-01', periods=365)
rows = []

for city in cities:
    for date in dates:
        rows.append({
            'timestamp': date,
            'city': city,
            'pm25': np.random.normal(75, 15),
            'pm10': np.random.normal(120, 25),
            'no2': np.random.normal(40, 10),
            'co': np.random.normal(0.9, 0.2)
        })

df = pd.DataFrame(rows)

# 4. Clean the data
df.dropna(inplace=True)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.columns = df.columns.str.lower().str.replace(' ', '_')

# 5. Save cleaned dataset
df.to_csv('pollution_monitoring/cleaned_pollution_data.csv', index=False)
df.head()

df
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("pollution_monitoring/cleaned_pollution_data.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Forecast for one city (e.g., Delhi)
city_df = df[df['city'] == 'Delhi'].copy()
city_df = city_df.sort_values('timestamp')

# Prophet formatting
df_prophet = city_df[['timestamp', 'pm25']].rename(columns={'timestamp': 'ds', 'pm25': 'y'})

# Model and forecast
model = Prophet()
model.fit(df_prophet)
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Plot forecast
model.plot(forecast)
plt.title("30-Day PM2.5 Forecast - Delhi")
plt.grid(True)
plt.show()
```

```python
import spacy
from transformers import pipeline
import pandas as pd

# Example synthetic news dataset
news = pd.DataFrame({
    'news_text': [
        "Delhi faces hazardous air conditions as PM2.5 soars.",
        "Mumbai launches electric bus initiative to cut CO levels.",
        "Chennai sees sharp rise in NO2 from industrial activity.",
        "Bangalore air quality improves after rain.",
        "Kolkata ramps up air pollution monitoring stations."
    ]
})

# NLP processing
nlp = spacy.load("en_core_web_sm")
summarizer = pipeline("summarization")

results = []

for text in news['news_text']:
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    summary = summarizer(text, max_length=50, min_length=25, do_sample=False)[0]['summary_text']
    results.append({'text': text, 'entities': entities, 'summary': summary})

alerts_df = pd.DataFrame(results)
alerts_df.to_csv("pollution_monitoring/news_alerts_summary.csv", index=False)
alerts_df
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import os

# Load data
df = pd.read_csv('pollution_monitoring/cleaned_pollution_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Output folder
os.makedirs('pollution_monitoring/ml_predictions', exist_ok=True)

# Pollutants to predict
pollutants = ['pm25', 'pm10', 'no2', 'co']

# Lag feature function
def create_multi_lags(df, cols, lags=3):
    for col in cols:
        for i in range(1, lags + 1):
            df[f'{col}_lag_{i}'] = df[col].shift(i)
    return df.dropna()

# Store RMSE scores
rmse_results = []

# Loop over all cities
for city in df['city'].unique():
    print(f"\nTraining for city: {city}")
    
    city_df = df[df['city'] == city].copy().sort_values('timestamp')
    city_df_lagged = create_multi_lags(city_df, pollutants)

    if city_df_lagged.shape[0] < 100:
        print(f"  Skipping {city} (not enough data after lagging)")
        continue

    # Input features & targets
    X = city_df_lagged[[f'{col}_lag_{i}' for col in pollutants for i in range(1, 4)]]
    y = city_df_lagged[pollutants]

    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    # Model
    model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Save predictions
    result_df = y_test.reset_index(drop=True).copy()
    result_df[['pm25_pred', 'pm10_pred', 'no2_pred', 'co_pred']] = y_pred
    result_df.to_csv(f'pollution_monitoring/ml_predictions/{city}_predictions.csv', index=False)

    # Evaluate
    for i, pollutant in enumerate(pollutants):
        rmse = np.sqrt(mean_squared_error(y_test[pollutant], y_pred[:, i]))
        rmse_results.append({'city': city, 'pollutant': pollutant, 'rmse': round(rmse, 2)})

print("\nâœ… Forecasts and evaluations complete.")

# Save RMSE summary
pd.DataFrame(rmse_results).to_csv("pollution_monitoring/ml_predictions/model_metrics_summary.csv", index=False)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Define regression accuracy
def regression_accuracy(y_true, y_pred, tolerance=10):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    within_range = np.abs(y_true - y_pred) <= tolerance
    return np.mean(within_range) * 100

# Define MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    nonzero = y_true != 0
    return np.mean(np.abs((y_true[nonzero] - y_pred[nonzero]) / y_true[nonzero])) * 100

# Loop over cities
for city in df['city'].unique():
    print(f"\nTraining model for city: {city}")
    
    city_df = df[df['city'] == city].copy().sort_values('timestamp')
    city_df_lagged = create_multi_lags(city_df, pollutants)

    if city_df_lagged.shape[0] < 100:
        print(f"Skipping {city} (not enough data after lagging)")
        continue

    X = city_df_lagged[[f'{col}_lag_{i}' for col in pollutants for i in range(1, 4)]]
    y = city_df_lagged[pollutants]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"Evaluation Metrics for {city}:\n")
    for i, pollutant in enumerate(pollutants):
        true = y_test[pollutant].values
        pred = y_pred[:, i]

        rmse = np.sqrt(mean_squared_error(true, pred))
        mae = mean_absolute_error(true, pred)
        r2 = r2_score(true, pred)
        mape = mean_absolute_percentage_error(true, pred)
        acc = regression_accuracy(true, pred, tolerance=10)

        print(f"â–¶ {pollutant.upper()}")
        print(f"   RMSE     : {rmse:.2f}")
        print(f"   MAE      : {mae:.2f}")
        print(f"   RÂ²       : {r2:.3f}")
        print(f"   MAPE     : {mape:.2f}%")
        print(f"   Accuracy : {acc:.2f}% (within Â±10 units)\n")
```
